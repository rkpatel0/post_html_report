```python
import regression.database as db
import regression.report_gen as gen
import webgen.report as reb
```

```python
# -----------------------------------------------------------------------------
# Configurable Parameters
# ------------------------------------------------------------------------------
CHANNEL = 'A'                       # 'A' or 'B'
EVKTYPE = 'EVK1900'                 # EVK1900 / EVK900 / EVK2400 / PMUTEST
TESTTYPE = 'CHAR'                   # CHAR / PMU / SOAK
REPORT_EXT = '116A_TTT_SKY23_NOD4dBATTN'
SAVE_PATH = r'C:\Users\Rishi.Patel\Desktop\\'   # Location where results should be saved
PATH_TO_DATA =  r'\\lightspeed1\workspace\work_rpatel\regression\result\CHAR\EVK1900\\'  # typically leave '' if using the default database

# Files to Read
logfiles = [
            'TT_9620150727T150906.csv',
#           'SSS_11720150114T120049.csv',    # use to append additional .csv files
            ]

analysis = {
            'filter': ['s_Waveform', 's_Backoff_dB', 's_Fw'],    # Add param to overwrite default
#            'sweep': ['s_Vdd_port_1_Vs', 's_Temperature_degC', 'EVBNUM'],
#            'repeat': ['r_RepeatNumWithoutReset', 's_Freq_MHz'],
#            'xaxis': 's_Freq_MHz',
            }
```

```python
# -----------------------------------------------------------------------------
# Generate Objects based on test type
# ------------------------------------------------------------------------------
if TESTTYPE == 'CHAR':
    oGen = gen.Char(config=analysis, chan=CHANNEL)
elif TESTTYPE == 'PMU':
    oGen = gen.Pmu(config=analysis, chan=CHANNEL)
elif TESTTYPE == 'SOAK':
    oGen = gen.Soak(config=analysis, chan=CHANNEL)
else:
    raise IndexError('Test Type Does not Exist ' + TESTTYPE)
    
oData = db.Reader()
oWeb = reb.MaximIntegrated({'save_path': SAVE_PATH})
```

```python
# -----------------------------------------------------------------------------
# Read Data, Process and Generate Report
# -----------------------------------------------------------------------------
data = oData.fetch_datalogs(logfiles, EVK=EVKTYPE, TEST=oGen.TESTTYPE, path=PATH_TO_DATA)
pages = oGen.data_to_report(data.reset_index(), oData.info)
oWeb.generate_report(pages, name=TESTTYPE + '_' + EVKTYPE + '_' + REPORT_EXT)
```
